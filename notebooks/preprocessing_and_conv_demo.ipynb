# Preprocessing and Convolution Demo

This notebook is used only for **sanity checking and visualization** of preprocessing and manual convolution logic.

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Load and Display Image

```python

img = cv2.imread('../dataset/with_mask/with_mask_1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
```

## Preprocessing

```python

gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
gray = cv2.resize(gray, (64, 64))

plt.imshow(gray, cmap='gray')
plt.title("Grayscale + Resized (64x64)")
plt.axis('off')
```

## Manual Convolution

```python

kernel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

def conv2d(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape
    out = np.zeros((h-kh+1, w-kw+1))

    for i in range(h-kh+1):
        for j in range(w-kw+1):
            out[i, j] = np.sum(image[i:i+kh, j:j+kw] * kernel)
    return out

feature_map = conv2d(gray, kernel)

plt.imshow(feature_map, cmap='gray')
plt.title("Feature Map (Edge Detection)")
plt.axis('off')
```
